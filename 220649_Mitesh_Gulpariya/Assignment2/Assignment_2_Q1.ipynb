```python
# Extract feature and target matrices
features = data[['Age', 'EstimatedSalary']]
targets = data['Purchased']

# Split the data into training and test sets with a 70:30 ratio
X_train, X_test, y_train, y_test = train_test_split(features, targets, test_size=0.3, random_state=42)
```

```python
# sigmoid function
def sigmoid(x):
    return (1/(1+np.exp(-x)))
```

```python
# Logistic regression from scratch
class LogisticRegression():

    def __init__(self, L=0.001, n_iters=1000):
        self.L = L
        self.n_iters = n_iters
        self.weights = None
        self.bias = None

    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.bias = 0

        for _ in range(self.n_iters):
            linear_pred = np.dot(X, self.weights) + self.bias
            linear_pred = np.array(linear_pred, dtype = np.longfloat)
            predictions = sigmoid(linear_pred)

            dw = (1/n_samples) * np.dot(X.T, (predictions - y))
            db = (1/n_samples) * np.sum(predictions-y)

            self.weights = self.weights - self.L*dw
            self.bias = self.bias - self.L*db


    def predict(self, X):
        linear_pred = np.dot(X, self.weights) + self.bias
        y_pred = sigmoid(linear_pred)
        class_pred = [0 if y<=0.5 else 1 for y in y_pred]
        return class_pred
```

```python
clf = LogisticRegression(L=0.01)
clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)

def accuracy(y_pred, y_test):
    return np.sum(y_pred==y_test)/len(y_test)

acc = accuracy(y_pred, y_test)
print(acc)
```

```python
scaler = StandardScaler()

X_train1 = scaler.fit_transform(X_train)

X_test1  = scaler.transform(X_test)
```

```python
clf.fit(X_train1,y_train)
y_pred = clf.predict(X_test1)
acc = accuracy(y_pred, y_test)
print(acc)
```

```python
scaler1 = MinMaxScaler()

X_train2 = scaler1.fit_transform(X_train)

X_test2  = scaler1.transform(X_test)
```

```python
clf.fit(X_train2,y_train)
y_pred = clf.predict(X_test2)
acc = accuracy(y_pred, y_test)
print(acc)
```
